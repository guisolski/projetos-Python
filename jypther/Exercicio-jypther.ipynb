```python
def somaV(v1,v2):
    vetor = []
    if len(v1) != len(v2):
        return False
    for i in range(len(v1)):
        vetor.append(v1[i]+v2[i])
    return tuple(vetor)
def diferencaV(v1,v2):
    vetor = []
    if len(v1) != len(v2):
        return False
    for i in range(len(v1)):
        vetor.append(v1[i]+(-1*v2[i]))
    return tuple(vetor)
```

```python
def moduloV(v):
    if type(v) != tuple:
      return False
    c = 0
    for i in range(len(v)):
        c += (v[i]**2)
    return (c**(1/2))
def plot(v1,v2):
    plt.plot((v1[0],v2[0]),(v1[1],v2[1]))
def cos(v,modulo):
    if type(v) != tuple:
        return False
    vetor = []
    for i in range(len(v)):
        vetor.append(v[i]/modulo)
    return tuple(vetor)
    
```

```python
origem =(0,0)
print("Vetor (5,6) tem modulo igual:", moduloV((5,6)),"  e tem os cosseos em relação a x e y especificamente: ",cos((5,6),moduloV((5,6))) )
plot(origem,(5,6))
print("Vetor (5,-5) tem modulo igual:", moduloV((5,-5)),"  e tem os cosseos em relação a x e y especificamente: ",cos((5,-5),moduloV((5,-5))) )
plot(origem,(5,-5))
print("Vetor (0,7) tem modulo igual:", moduloV((0,7)),"  e tem os cosseos em relação a x e y especificamente: ",cos((0,7),moduloV((0,7))) )
plot(origem,(0,7))
print("Vetor (-7,0) tem modulo igual:", moduloV((-7,0)),"  e tem os cosseos em relação a x e y especificamente: ",cos((-7,0),moduloV((-7,0))) )
plot(origem,(-7,0))
print("Vetor (-5,3) tem modulo igual:", moduloV((-5,3)),"  e tem os cosseos em relação a x e y especificamente: ",cos((-5,3),moduloV((-5,3))) )
plot(origem,(-5,3))
```

```python
def unitario(v,modulo):
    vetor = []
    for i in range(len(v)): vetor.append( v[i]/modulo)
    return tuple(vetor)
def prodEscalar(v,k):
    vetor = []
    for i in range(len(v)):
            vetor.append(v[i]*k)
    return tuple(vetor)
```

```python
a = (-2,0,7)
b = (4,1,3)
print("Unitario de A: ", unitario(a,moduloV(a)))
print("Unitario de B: ", unitario(b,moduloV(b)))
print("Modulo de A: ", moduloV(a))
print("Modulo de B: ", moduloV(b))
print("c = a-2b: ", diferencaV(a,prodEscalar(b,2)))
print("d = 3a: ", prodEscalar(a,3))
print("e = 2a + 3b: ", somaV(prodEscalar(a,2),prodEscalar(b,3)))
```

```python
def produtoEscalar(v1,v2):
    if len(v1) != len(v2): return False
    if type(v1) != tuple or type(v2) != tuple: return False
    resultado = 0
    for i in range(len(v1)): resultado += v1[i]*v2[i]
    return resultado
```

```python
i = (1,0,0)
j = (0,1,0)
k = (0,0,1)
print(produtoEscalar(i,i))
print(produtoEscalar(i,j))
print(produtoEscalar(i,k))
print(produtoEscalar(j,j))
print(produtoEscalar(j,k))
print(produtoEscalar(k,k))
```

## Exercicio 16

```python
def somaV(v1,v2):
    vetor = []
    if len(v1) != len(v2):
        return False
    for i in range(len(v1)):
        vetor.append(v1[i]+v2[i])
    return tuple(vetor)
def prodEscalar(v,k):
    vetor = []
    for i in range(len(v)):
            vetor.append(v[i]*k)
    return tuple(vetor)

def pU(p0,t,u):
    if type(p0) != tuple or type(t) != tuple:
        return False
    if len(p0) != len(t):
        return False
    return tuple((somaV(prodEscalar(t,u),p0)))
```

```python
origem =(0,0,0)
a = (6,-1,-2)
b = (3,2,4)
c = (7,0,2)

fig = plt.figure()
ax = plt.axes(projection='3d')

u =  np.linspace(-1.0,2.0,8)

Ponto = pU(b,a,u)


plot(origem,a)
plot(origem,b)
plot(Ponto[0],Ponto[1])
```
