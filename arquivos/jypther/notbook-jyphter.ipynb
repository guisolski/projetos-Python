```python
def fatorial(valor):
    if valor <= 1:
        return 1
    else:
        return valor*fatorial(valor-1)
    
def somatoria(valor):
    soma = 0
    for i in range(valor):
        soma += 1/fatorial(i)
    return soma

def lista(valor):
    Lista = []
    for i in range(1,valor+1):
        Lista.append(1/i)
    return Lista

def listaSimples(valor):
    return [1/i for i in range(1,valor+1)]

def listaSoma(valor):
    return sum(valor)

def listaLambda(valor):
    import functools
    soma = lambda x,y : x + y
    return functools.reduce(soma,valor)
```

```python
def somaV(v1,v2):
    vetor = []
    if len(v1) != len(v2):
        return False
    for i in range(len(v1)):
        vetor.append(v1[i]+v2[i])
    return tuple(vetor)
def diferencaV(v1,v2):
    vetor = []
    if len(v1) != len(v2):
        return False
    for i in range(len(v1)):
        vetor.append(v1[i]+(-1*v2[i]))
    return tuple(vetor)

def prodEscalar(v,k):
    vetor = []
    for i in range(len(v)):
            vetor.append(v[i]*k)
    return tuple(vetor)
def moduloV(v):
    if type(v) != tuple:
      return False
    c = 0
    for i in range(len(v)):
        c += (v[i]**2)
    return (c**(1/2))
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
```

```python
def Casinha(x,y):
    dx = 1
    dy = 1
    vertsT = [
        (x, y+dy), # left, bottom
        (x, y+dy+dy), # left, top
        (x+dx, y+dy), # right, top
        (x+dx, y+dy), # right, bottom
        (x, y), # ignored
    ]

    codesT = [Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.CLOSEPOLY,]
    vertsRect = [
        (x, y), # left, bottom
        (x, y+dy), # left, top
        (x+dx, y+dy), # right, top
        (x+dx, y), # right, bottom
        (x, y), # ignored
    ]

    codesRect = [Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.CLOSEPOLY,]

    pathT = Path(vertsT, codesT)
    pathRect = Path(vertsRect, codesRect)
          
    patchT = patches.PathPatch(pathT, facecolor='green', lw=1)
    patchRect = patches.PathPatch(pathRect, facecolor='orange', lw=1)
    retorna = (patchT,patchRect)
    return retorna
```

```python
import math
fig, ax = plt.subplots()
plt.xlim(-11,11)
plt.ylim(0,10)

f = lambda x: math.sin(x)

x1 = np.linspace(-10, 100, 100)

for px in x1:
    casa = Casinha(px,f(px))
    ax.add_patch(casa[0])
    ax.add_patch(casa[1])


ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
 
#fig.savefig("circulosSobreFuncaoQualquer.png")
plt.show()
```
